```python
import numpy as np
def ticker_find(xml_dict, ticker):
    return xml_dict[ticker]

def calc_avg_open(csv_data, ticker):
    return csv_data.loc[csv_data['Symbol'] == ticker, 'Open'].mean()

def vwap(csv_data, ticker):
    frame = pd.DataFrame()
    frame = frame.assign(High = csv_data.loc[csv_data['Symbol'] == ticker, 'High'])
    frame = frame.assign(Low = csv_data.loc[csv_data['Symbol'] == ticker, 'Low'])
    frame = frame.assign(Close = csv_data.loc[csv_data['Symbol'] == ticker, 'Close'])
    frame = frame.assign(Volume = csv_data.loc[csv_data['Symbol'] == ticker, 'Volume'])
    frame['AveragePrice'] = frame[['High', 'Low', 'Close']].apply(sum, axis=1)
    frame['AveragePrice'] = frame['AveragePrice'] / 3
    frame['AveragePriceVolume'] = frame['AveragePrice'] * frame['Volume']
    Numerator =  frame[['AveragePriceVolume']].apply(sum, axis=0)
    Denominator = frame[['Volume']].apply(sum, axis=0)
    VolumeWeightedAverage = np.float64(Numerator) / np.float64(Denominator)
    return (VolumeWeightedAverage)
    
```

```python
import pandas as pd

df = pd.read_csv("SP500_ind.csv")
df.head(5)
```

```python
import untangle
input_file = "http://www.cs.odu.edu/~sampath/courses/f18/cs795/files/data/SP500_symbols.xml"
obj = untangle.parse(input_file)
obj
```

```python
xml_dict = {}
for symbol in obj.symbols.symbol:
    xml_dict[symbol['ticker']] = symbol['name']
    
    
```

```python
ticker = df['Symbol'].unique().tolist()
```

```python
for l in ticker:
    
    try:
        name = ticker_find(xml_dict, l)
        average_open = calc_avg_open(df, l)
        vwap_values = vwap(df, l)
        print(name + ' ' + str(average_open) + ' ' + str(vwap_values))
    
    except:
        name = 'No data in SP500'
        average_open = calc_avg_open(df, l)
        vwap_values = vwap(df, l)

        print(name + ' ' + str(average_open) + ' ' + str(vwap_values))
        continue
```
